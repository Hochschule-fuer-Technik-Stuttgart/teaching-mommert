## Loading Image Data

The easiest way to deal with regular RGB image data in Python is using the [Python Image Library](https://pillow.readthedocs.io/en/stable/index.html) (PIL or pillow). In the following, we will read an image from a url, which is somewhat more complicated than reading an image from file:

```python
from PIL import Image
import requests
from io import BytesIO

response = requests.get('https://www.hft-stuttgart.de/fileadmin/Dateien/Marketing/_processed_/9/a/csm_studium-an-der-hochschule-fuer-technik-stuttgart-luftbild_10673cfebe.jpg')
img = Image.open(BytesIO(response.content))
```

Let's have a look at `img`, which is an `Image` object, and its properties. Jupyter Notebooks provide a nice feature with regard to `Image` objects: your Notebook environment will notice that it contains an image simply display it:

```python
img
```

`Image` objects contain raster image data, which can be easily converted into a NumPy array:

```python
a = np.array(img)
a.shape
```

The shape of this array tells us the image dimensions: the image has a **height** of 635 pixels, a **width** of 952 pixels and a **depth** of 3 channels. The output of shape is therefore: `(height, width, depth)`. Since `a` is an array, we can apply all array operations. 

Before we do so, let's figure out how we can plot this image:

```python
plt.imshow(a)
```

That's easy. In this case, the image is accompanied by a **coordinate system**. The important thing about image coordinate systems in Python is that pixel coordinates are provided in rows and columns. 

Let's use this knowledge and crop that part of the image that contains that one hirise building close to the center of the image:

```python
plt.imshow(a[150:300, 400:500])
```

We can also extract **individual pixel values** by simply addressing a single pixel:

```python
a[600, 800]
```

We learned how to extract spatial regions from the image. In the same way, we can extract **spectral information**. Let's extract the R, G and B data into separate arrays:

```python
r = a[:, :, 0]
g = a[:, :, 1]
b = a[:, :, 2]
```

Let's plot the three bands separately next to each other:

```python
f, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(r, cmap='Greys_r', vmin=0, vmax=255)
ax[0].set_title('R')

ax[1].imshow(g, cmap='Greys_r', vmin=0, vmax=255)
ax[1].set_title('G')

ax[2].imshow(b, cmap='Greys_r', vmin=0, vmax=255)
ax[2].set_title('B')
```

If we have each channel as a separate array, we can combine them together in the way that is required for the imshow function. Let's do this but we will deliberately swap the red and blue bands:

```python
plt.imshow(np.dstack([b, g, r]))
```

## Modifying data

Since images are just arrays, we can easily modify them. For instance, we can modify the pixel values with arithmetic operations:

```python
f, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(a+50)
ax[1].imshow(a-50)
```

Note the funny colors! 

**Question**: Where do the funny colors come from?

Let's change the data type of our array and try again:

```python
a = a.astype(int)

f, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(a+50)
ax[1].imshow(a-50)
```

This looks better: the first image has an increased brightness, the second image has a decreased brightness. However, we receive a warning: `imshow` expects all pixel values to be within the range [0, 255] (for integer values) or [0,1] (for floats), which is not the case because of our modifications. Anyway, we can ignore the warning, as values outside this range will be ignored.

In order to adjust the image brightness, we can scale pixel values by multiplying our pixel array data with a constant value. This is called **linear scaling**.

To better analyze the brightness distribution of an image, we can use **histograms**.

In the following, we plot the distribution of green pixel values:

```python
green_hist, bins = np.histogram(a[:, :, 1].reshape(-1), bins=np.arange(0, 255, 5, dtype=int), density=True)

plt.plot(bins[:-1], green_hist, c='green')
```
